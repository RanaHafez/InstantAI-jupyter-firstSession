# TASK 1: read all sheets of excel file using pandas

```python
import pandas as pd
```

### First Way using a loop (explicit way)

```python
file = pd.ExcelFile('TestforGettingAllSheets.xlsx')
```

```python

all_sheets_dic = {}
print(file.sheet_names)
for sheet_name in file.sheet_names:
    all_sheets_dic[sheet_name] = file.parse(sheet_name)
```

```python
type(all_sheets_dic['Sheet1'])
```

### different way more implicit with out specifying a sheet name

```python
df = pd.read_excel('TestforGettingAllSheets.xlsx', sheet_name=None )
```

#### by not specifying the sheet name all sheets will be in a dictionary with each being od dataframe

```python
type(df['Sheet1'])
```

# TASK 2: how to append to a existing csv file and save the sheet

```python
import csv
```

```python
with open('csv_file.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    field = ["Name", "Age"]

    writer.writerow(field)
    writer.writerow(["koko", 40])
    writer.writerow(["Soso", 23])
    writer.writerow(["Lolo", 50])
```

```python
df['Sheet1']
```

### the answer is to specify the mode to be 'a' which does mean append to an existing file

```python
df['Sheet1'].to_csv('csv_file.csv', mode='a', index=False, header=False)
```
